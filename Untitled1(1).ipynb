```python
data = {
    "Name": ["Menna", "Ahmed", "Omar", "Mona"],
    "Age": [20, 21, 19, 22],
    "Math_Score": [85, 67, 90, 73],
    "Gender": ["M", "F", "M", "F"]
}

df = pd.DataFrame(data)

print(df.info())
print(df.describe())
```

```python
print(df["Name"])
print(df.loc[0])
print(df.iloc[1])

filtered = df[df["Math_Score"] > 70]
print(filtered)
```

```python
df.loc[2, "Math_Score"] = None

df = df.fillna(0)
df = df.drop_duplicates()
df = df.drop("Age", axis=1)

print(df)
```

```python
def result(score):
    return "Pass" if score >= 70 else "Fail"

df["Result"] = df["Math_Score"].apply(result)

dummies = pd.get_dummies(df["Gender"])
print(dummies)
```

```python
df2 = pd.DataFrame({
    "Name": ["Khaled", "Nour"],
    "Math_Score": [88, 92],
    "Gender": ["M", "F"],
    "Result": ["Pass", "Pass"]
})

combined = pd.concat([df, df2])
print(combined)

pivot = combined.pivot_table(values="Math_Score", index="Gender", aggfunc="mean")
print(pivot)
```
